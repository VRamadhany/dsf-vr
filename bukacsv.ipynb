```python
import matplotlib
import pandas as pd
import csv as csv
import matplotlib.pyplot as plt
import os
##########  Buka CSV
bukacsv = pd.read_csv("pokemon.csv")
count = bukacsv['Total'].value_counts()
ranked = bukacsv.sort_values(by='Total', ascending=False)
bukacsv = ranked.head(10)

##########  Tampilkan
bukacsv = bukacsv.dropna(subset=['Name'])
plt.barh(bukacsv['Name'], bukacsv['Total'], color='blue', height=0.5)
plt.title('Rank Berdasarkan Total Power')
plt.xlabel('Total')
plt.ylabel('Name')

plt.show()
```
